### Data Visualization With Seaborn
Seaborn is a Python visualization library based on Matplotlib that provides a high-level interface for drawing attractive and informative statistical graphics. Seaborn helps in creating complex visualizations with just a few lines of code.

```python
import seaborn as sns 
```

```python
### Basic Plotting with Seaborn
tips_Dataset = sns.load_dataset('tips')
tips_Dataset
```

```python
# Create a scatter plot

import matplotlib.pyplot as plt

sns.scatterplot(x='total_bill' , y='tip' , color='teal',data=tips_Dataset)
plt.title("Scatter plot of Total Bill v/s Tip")
plt.show()
```

```python
## Line Plot

sns.lineplot(x='size',y='total_bill',data=tips_Dataset)
plt.title("Line plot of Total Bill v/s Size")
plt.show()
```

```python
## Categorical Variable
sns.barplot(x='day',y='total_bill' , data=tips_Dataset)
plt.title("The bar plot of Total Bill v/s Size")
plt.show()
```

```python
## Box Plot
sns.boxplot(x='day',y='total_bill',data = tips_Dataset)
```

```python
# Violin Plot
sns.violinplot(x='day',y='total_bill' , data = tips_Dataset)
plt.title("Voilin Plot")
plt.show()
```

```python
# Histograms
sns.histplot(tips_Dataset['total_bill'] ,bins = 10 , kde = True) 
# kds = Kernel Density Estimator
```

```python
# KDE plot
sns.kdeplot(tips_Dataset['total_bill'] , color = 'red' , shade = True)
```

```python
# Pairplot
sns.pairplot(tips_Dataset)
```

```python
# Heat Map
corr = tips_Dataset[['total_bill','tip','size']].corr()
corr
```

```python
sns.heatmap(corr,annot=True,cmap='coolwarm')
```
